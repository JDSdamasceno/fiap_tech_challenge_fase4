```python
import yfinance as yf

# import pandas as pd
# import matplotlib.pyplot as plt
```

```python
ativo = "ITUB3.SA"

data_inicial = "2022-01-01"
data_final = "2025-12-31"

tabela_cotacao = yf.download(
    ativo, start=data_inicial, end=data_final, progress=False
)
tabela_cotacao = tabela_cotacao.xs(ativo, level="Ticker", axis=1)
tabela_cotacao["adjust_close"] = tabela_cotacao["Close"].pct_change()
tabela_cotacao.columns.name = None
tabela_cotacao = tabela_cotacao.reset_index()

tabela_cotacao
```
